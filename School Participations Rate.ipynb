```python
# Library used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

## 1. Data Preparation

First we will import the dataset into the notebook

```python
# Import School Partitipations Rate
path = './Angka Partisipasi Sekolah.xlsx'
df = pd.read_excel(path)
display(df.head())
```

```python
# Import GeoJSON Data
path = './gadm36_IDN_1.json'
df_geo = gpd.read_file(path)
display(df_geo.head())
```

This "gpd.read_file" function will directly create a GeoDataFrame class.
Furthermore We can directly plot this GeoDataFrame with simple plot function

```python
# Check what kind of object districts is
print('df_geo data type: {}'.format(type(df_geo)))

# Plot GeoDataFrame
print('\nPlotting GeoDataFrame: ')
df_geo.plot()
```

## 2. Merge Datase

In this step, we will join the dataframe that includes School Partitipations Rate data and GeoDataFrame. This will give us a GeoDataframe that have School Participations Rate values on it.

```python
# join the geodataframe with the csv dataframe
df_join = df_geo.merge(df, how='inner', left_on="NAME_1", right_on="Province")
df_join = df_join[['Province','7-12', '13-15', '16-18', '19-24','geometry']]
df_join.head()
```

## 3. Create a map

To create a choroplet we need to specify which values we will show on a map and prepare the matplotlib map

```python
# set a variable that will call whatever column we want to visualise on the map
values = '19-24'

# set the value range for the choropleth
vmin, vmax = df_join[values].min(), df_join[values].max()

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(50, 70))

# create map
df_join.plot(column=values, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')
```

Now we have a map with shaded color that represent values on it. Next we will do some customisation for this map.

## Customise Map

First, we will give more detail on a map. We will adds Map title, color bar and annotation on the bottom of the map. We also will remove the axis on the map.

```python
# set a variable that will call whatever column we want to visualise on the map
values = '19-24'

# set the value range for the choropleth
# vmin, vmax = df_join[values].min(), df_join[values].max()
vmin, vmax = 0,100

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(30, 10))

# remove the axis
ax.axis('off')

# add a title
title = 'School Participations Rate for Age {}'.format(values)
ax.set_title(title, fontdict={'fontsize': '25', 'fontweight' : '3'})

# create an annotation for the data source
ax.annotate('Source: Badan Pusat Statistik Indonesia',xy=(0.1, .08),  xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', fontsize=12 ,color='#555555')

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))

# add the colorbar to the figure
cbar = fig.colorbar(sm)

# create map
df_join.plot(column=values, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8',norm=plt.Normalize(vmin=vmin, vmax=vmax))
```

Lastly we will overlay province name on the map to give user more information

## Create All Map

Now we will make a function from above. So we can plot all data in a dataframe.

```python
def make_plot(geo_df,values):
    # set a variable that will call whatever column we want to visualise on the map
    values = geo_df.columns[values]

    # set the value range for the choropleth
    vmin, vmax = 0,100

    # create figure and axes for Matplotlib
    fig, ax = plt.subplots(1, figsize=(30, 10))

    # remove the axis
    ax.axis('off')

    # add a title
    title = 'School Participations Rate for Age {}'.format(values)
    ax.set_title(title, fontdict={'fontsize': '25', 'fontweight' : '3'})

    # create an annotation for the data source
    ax.annotate('Source: Badan Pusat Statistik Indonesia',xy=(0.1, .08),  xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', fontsize=12 ,color='#555555')

    # Create colorbar as a legend
    sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))

    # add the colorbar to the figure
    cbar = fig.colorbar(sm)

    # create map
    geo_df.plot(column=values, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8',norm=plt.Normalize(vmin=vmin, vmax=vmax))
    
    # Add Labels
    geo_df['coords'] = geo_df['geometry'].apply(lambda x: x.representative_point().coords[:])
    geo_df['coords'] = [coords[0] for coords in geo_df['coords']]
    for idx, row in df_join.iterrows():
        plt.annotate(s=row['Province'], xy=row['coords'],horizontalalignment='center', color = 'Red')
```

```python
for i in range(1,5):
    make_plot(df_join,i)
```
